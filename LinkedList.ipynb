#### Create Linked List

```python
## Node Class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        

## List Class
class LinkedList:
    def __init__(self):
        self.head = None
        
    ## This function add a node to the list    
    def addNode(self, item):
        list_node = Node(item)
        
        if self.head == None:
            self.head = list_node
        else:
            list_node.next = self.head
            self.head = list_node
   
    ## This function deiplays the content of current List
    def displayList(self):
        if self.head == None:
            print("List is empty")
        else:
            current_node = self.head
            while current_node:
                #print("{}\t".format(current_node.data)),
                print(current_node.data, end = " ")
                current_node = current_node.next
    
    ## This function reverse the current LinkedList
    def reverseList(self):
        curr_node = self.head
        prev_node = None
       
        while curr_node:
            
            #print('\nPrev: {} == Curr: {} == Next: {}'.format(prev_node, curr_node, next_node))
            next_node = curr_node.next
            curr_node.next = prev_node
            prev_node = curr_node
            curr_node = next_node
            
        self.head = prev_node
        
    ## This function create one list by adding the 2 lists of same length
    def addTwoList(self, second_list):
        current1 = self.head
        current2 = second_list.head
        
        while current1:
            current1.data = current1.data + current2.data
            current1 = current1.next
            current2 = current2.next 
            
```

```python
#### Perform List operations
my_list = LinkedList()

## Display List
my_list.displayList()

## Add Nodes
my_list.addNode(20)
my_list.addNode(30)
my_list.addNode(40)

## Display List
print("\n")
my_list.displayList()

## Reverse List
my_list.reverseList()
print("\n")
my_list.displayList()

## Create another list
your_list = LinkedList()

## Add Nodes
your_list.addNode(50)
your_list.addNode(80)
your_list.addNode(10)

## Display List
print("\n")
your_list.displayList()

## print("Add Two lists")
my_list.addTwoList(your_list)

## Display List
print("\n")
my_list.displayList()

## Reverse the Resultant List
my_list.reverseList()

## Display List
print("\n")
my_list.displayList()
```
